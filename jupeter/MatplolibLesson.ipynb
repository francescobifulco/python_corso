### ESEMPIO PLOT BASE

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

### MARKERS

```python

x = [0, 1, 2, 3, 4]
y = [0, 1, 2, 3, 4]


plt.figure(figsize=(3, 3))
plt.title("My First Plot")
plt.xlabel("AsseX")
plt.ylabel("AsseY")
plt.grid(True)


plt.plot(x, y, marker= 'o', ms=15 , mec = 'r', mfc='y')
plt.show()
```

### LINESTYLE

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle='-', c="#5D08C5", linewidth='10')

plt.show()
```

### MULTILINE

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
x = [0,1,2,3]

plt.plot(y1, marker='o', ls='--')
plt.plot(y2, marker='*', ls='-.')

plt.fill_between(x, y2, y1, color='green', alpha=0.5)

plt.show()
```

### LABELS AND TITLE

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 320])

font1 = {'family':'serif', 'color':'blue', 'size':20}
font2 = {'family':'serif', 'color':'red', 'size':15}

plt.title("My Plot", fontdict=font1, loc='right')
plt.xlabel("AsseX", fontdict=font2, loc='left')
plt.ylabel("AsseY", fontdict=font2, loc='bottom')

# plt.plot(x, y)
# plt.show()
```

### GRID

```python

plt.grid(axis = 'y', ls = '--', linewidth=1.8, c='y')

plt.plot(x, y)
plt.show()
```

### SUBPLOT

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.figure(figsize=(16, 4))

plt.subplot(1, 3, 1)
plt.plot(x, y)
plt.title("PLOT1")
plt.xlabel("AsseX1")
plt.ylabel("AsseY1")

w = np.array([0,1,2,3])
z = np.array([10, 20, 30, 40])

plt.subplot(1, 3, 2) # righe totali, colonne totali, posizione del plot
plt.plot(w, z)
plt.title("PLOT2")
plt.xlabel("AsseX2")
plt.ylabel("AsseY2")

u = np.array([0,1,2,3])
v = np.array([100, 20, 350, 40])

plt.subplot(1, 3, 3) # righe totali, colonne totali, posizione del plot
plt.plot(u, v)
plt.title("PLOT3")
plt.xlabel("AsseX3")
plt.ylabel("AsseY3")

plt.suptitle("MySubPlots")

plt.show()
```

### SCATTER

```python
# Primo Plot Scatter
x = np.random.randint(0, 21, 10) # quanti anni ha la macchina
y = np.random.randint(80, 121, 10) # quanto è veloce

# colors = np.array(["red", "yellow", "blue"])
colors = np.linspace(0, 100, 10)
sizes = np.random.randint(200, 300, 10)

# plt.scatter(x, y, c="#74ff5b")
plt.scatter(x, y, c=colors, cmap='viridis', s=sizes, alpha=0.5)

plt.colorbar()

# Secondo Plot Scatter

x = np.random.randint(0, 21, 10) # quanti anni ha la macchina
y = np.random.randint(80, 121, 10) # quanto è veloce

# colors = np.array(["black", "pink", "brown"])
colors = np.linspace(0, 100, 10)
sizes = np.random.randint(50, 100, 10)

# plt.scatter(x, y, c="#ff00f2")
plt.scatter(x, y, c=colors, cmap='viridis', s=sizes)

plt.show()
```

### BAR

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)

plt.barh(x, y, color="#4eb2c1", height = 0.8)

# plt.show()

plt.subplot(1, 2, 2)

plt.bar(x, y, color="#14914c", width = 0.8)

plt.show()
```

### HISTOGRAM

```python
x = np.random.normal(170, 10, 250)

plt.hist(x, color="red", width=0.8)
plt.show()
```

### PIE CHARTS

```python
y = np.array([35, 25, 25, 15])
myLables = np.array(["Apples", "Bananas", "Cherries", "Oranges"])
myExplode = [0.2, 0.1, 0, 0]

plt.pie(y, labels=myLables, startangle=180, explode=myExplode, shadow=True)
plt.legend(title = "Four Fruits: ")
plt.show()
```
