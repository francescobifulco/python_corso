1. Creare un sistema che gestisca una biblioteca con classi Libro, Utente e Biblioteca. 
    - Ogni utente può prendere in prestito uno o più libri. 
    - Implementare metodi per aggiungere libri, registrare utenti e gestire prestiti.

```python
class Libro:
    def __init__(self, titolo, autore):
        self.titolo = titolo
        self.autore = autore
        self.disponibile = True  

class Utente:
    def __init__(self, nome, cognome):
        self.nome = nome
        self.cognome = cognome
        self.libri_in_prestito = [] 
         
class Biblioteca:
    def __init__(self):
        self.elenco_libri = []  
        self.elenco_utenti = []  

    def aggiungi_libro(self, libro):
        self.elenco_libri.append(libro)

    def registra_utente(self, utente):
        self.elenco_utenti.append(utente)

    def prendi_in_prestito(self, utente, libro):
        if libro in self.elenco_libri and libro.disponibile:
            utente.libri_in_prestito.append(libro)
            libro.disponibile = False
            print(f"Il libro '{libro.titolo}' è stato preso in prestito da {utente.nome} {utente.cognome}.")
        else:
            print(f"Il libro '{libro.titolo}' non è disponibile per il prestito.")

    def restituisci(self, utente, libro):
        if libro in utente.libri_in_prestito:
            utente.libri_in_prestito.remove(libro)
            libro.disponibile = True
            print(f"Il libro '{libro.titolo}' è stato restituito con successo.")
        else:
            print(f"Il libro '{libro.titolo}' non è stato trovato negli elenchi di prestito dell'utente.")



biblioteca = Biblioteca()

    
libro1 = Libro("Il signore degli anelli", "J.R.R. Tolkien")
libro2 = Libro("Harry Potter e la Pietra Filosofale", "J.K. Rowling")
biblioteca.aggiungi_libro(libro1)
biblioteca.aggiungi_libro(libro2)

    
utente1 = Utente("Mario", "Rossi")
utente2 = Utente("Anna", "Verdi")
biblioteca.registra_utente(utente1)
biblioteca.registra_utente(utente2)

    
biblioteca.prendi_in_prestito(utente1, libro1)
biblioteca.prendi_in_prestito(utente2, libro2)
biblioteca.restituisci(utente1, libro1)

   
print(f"Stato dei libri disponibili:")
for libro in biblioteca.elenco_libri:
    if libro.disponibile:
        print(f"- {libro.titolo} di {libro.autore}")
```

```python
class Veicolo:
    def __init__(self, marca, modello):
        self.marca = marca
        self.modello = modello
        self.km_percorsi = 0

    def sposta(self, km):
        self.km_percorsi += km
        print(f"{self.marca} {self.modello} si è spostato di {km} km. Totale: {self.km_percorsi} km.")


class Auto(Veicolo):
    def __init__(self, marca, modello, cilindrata):
        super().__init__(marca, modello)
        self.cilindrata = cilindrata

    def sposta(self, km):
        print(f"L'auto {self.marca} {self.modello} con cilindrata {self.cilindrata}cc sta viaggiando...")
        super().sposta(km)


class Bicicletta(Veicolo):
    def __init__(self, marca, modello, tipo_pedali):
        super().__init__(marca, modello)
        self.tipo_pedali = tipo_pedali

    def sposta(self, km):
        print(f"La bicicletta {self.marca} {self.modello} con pedali {self.tipo_pedali} sta pedalando...")
        super().sposta(km)




auto1 = Auto("Fiat", "Punto", 1200)
bici1 = Bicicletta("Bianchi", "Oltre XR4", "automatici")

auto1.sposta(50)
auto1.sposta(30)
bici1.sposta(10)
bici1.sposta(5)

print(f"\nChilometri finali percorsi:")
print(f"{auto1.marca} {auto1.modello}: {auto1.km_percorsi} km")
print(f"{bici1.marca} {bici1.modello}: {bici1.km_percorsi} km")
```

```python
class Animale:
    def __init__(self, nome):
        self.nome = nome

    def emetti_verso(self):
        print(f"{self.nome} emette un suono indefinito.")


class Leone(Animale):
    def emetti_verso(self):
        print(f"{self.nome} ruggisce: Roooar!")


class Elefante(Animale):
    def emetti_verso(self):
        print(f"{self.nome} barrisce: Poooooh!")


class Pappagallo(Animale):
    def emetti_verso(self):
        print(f"{self.nome} parla: Ciao! Sono un pappagallo!")


class Zoo:
    def __init__(self):
        self.animali = []

    def aggiungi_animale(self, animale):
        self.animali.append(animale)
        print(f"{animale.nome} è stato aggiunto allo zoo.")

    def rimuovi_animale(self, nome):
        for animale in self.animali:
            if animale.nome == nome:
                self.animali.remove(animale)
                print(f"{nome} è stato rimosso dallo zoo.")
                return
        print(f"Nessun animale con nome {nome} trovato.")

    def mostra_animali(self):
        print("\nAnimali presenti nello zoo:")
        for animale in self.animali:
            print(f"- {animale.nome} ({animale.__class__.__name__})")
            animale.emetti_verso()


zoo = Zoo()

leone = Leone("Simba")
elefante = Elefante("Dumbo")
pappagallo = Pappagallo("Polly")

zoo.aggiungi_animale(leone)
zoo.aggiungi_animale(elefante)
zoo.aggiungi_animale(pappagallo)

zoo.mostra_animali()

zoo.rimuovi_animale("Dumbo")
zoo.mostra_animali()
```
