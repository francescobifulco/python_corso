# SEABORN

#### Introduzione
Seaborn è una libreria Python utilizzata principalmente per la visualizzazione grafica di dati statistici.
Costruita sopra Matplotlib, fornisce stili predefiniti e palette di colori accattivanti per rendere i grafici più
attraenti e informativi . In questo tutorial impareremo ad usare Seaborn, dai concetti di base fino a quelli
avanzati, attraverso numerosi esempi pratici e grafici diversi . 


#### Iniziare con Seaborn
##### Installazione e dipendenze
Per installare Seaborn, si può utilizzare pip. Aprire un terminale e digitare:

`pip install seaborn`

Al termine dell’installazione dovreste vedere un messaggio di successo. Seaborn richiede le seguenti
dipendenze:
Python (versione 2.7 o 3.4+; Python 3 è raccomandato)
NumPy (per la manipolazione efficiente di array numerici)
SciPy (alcune funzioni statistiche avanzate)
Pandas (per gestire dataset e DataFrame)
Matplotlib (per le funzionalità grafiche di base su cui Seaborn si appoggia)

### SINGLE PLOT CONTEXT AND STYLE

```python
data = sns.load_dataset("iris")

def plot():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

plt.figure(figsize=(10,8))
sns.set_style("whitegrid")
sns.set_context("paper")

with sns.axes_style('darkgrid'):
    with sns.plotting_context("paper"):
        plt.subplot(2,1,1)
        plot()

plt.subplot(2,1,2)
plot()

plt.show()
```

### COLOR PALETTE

#### SET DEFAULT PALETTE

```python
sns.set_palette('vlag')
plt.subplot(2,1,1)
plot()

sns.set_palette('Accent')
plt.subplot(2,1,2)
plot()

plt.show()
```

### GRAFICI MULTIPLI

##### ADDAXES():PYPLOT

```python
def graph():
    sns.lineplot(x='sepal_length', y='sepal_width', data=data)
    
fig = plt.figure(figsize=(10,8))

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
graph()

ax2 = fig.add_axes([0.5,0.5,0.3,0.3])
graph()

plt.show()
```

#### SUBPLOT2GRID():PYPLOT

```python
axes1 = plt.subplot2grid((7, 1), (0, 0), rowspan=2, colspan=1) # Formato griglia , Start grafico (0,0) = in alto a sinistra, righe e colonne occupate
graph()

axes2 = plt.subplot2grid((7, 1), (2, 0), rowspan=2, colspan=1)
graph()

axes3 = plt.subplot2grid((7, 1), (4, 0), rowspan=2, colspan=1)
graph()

plt.show()
```

##### CATPLOT()

```python
dataTips = sns.load_dataset("tips")

sns.catplot(
    data = dataTips,
    x='day',
    y='tip',
    hue='sex',
    kind='swarm',
    palette="pastel",
    errorbar=None
)
plt.show()
```

#### GRAFICI DI REGRESSIONE

```python
#Realizzate un regplot che visualizzi la relazione tra la lunghezza (bill_length_mm) e la profondità (bill_depth_mm) del becco dal dataset penguins.
data = sns.load_dataset('penguins')
sns.regplot(x='bill_length_mm', y='bill_depth_mm', data = data)
```

```python
#Utilizzate lmplot per creare una regressione lineare separata per ogni isola (island) nel dataset penguins.
data = sns.load_dataset('penguins')
sns.lmplot(x='bill_length_mm', y='bill_depth_mm',hue='island', data = data)
plt.title('Regressione Lineare di Lunghezza vs. Profondità del Becco per Ogni Isola')
```

```python
#Realizzate un regplot che visualizzi la relazione tra il peso (carat) e il prezzo (price) dal dataset diamonds.
data = sns.load_dataset('diamonds')
sns.regplot(x='carat', y='price', data=data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#Utilizzate lmplot per creare una regressione lineare separata per il sesso (sex) nel dataset exercise, confrontando il battito cardiaco (pulse) e il tempo di esercizio (time).

data = sns.load_dataset('exercise')
data['time'] = data['time'].astype(str).str.replace(' min', '').astype(float)
sns.lmplot(x='time', y='pulse', hue='kind', data=data)

plt.title('Relazione tra Battito Cardiaco e Tempo di Esercizio, Separata per Tipo di Esercizio')
plt.xlabel('Tempo di Esercizio (minuti)')
plt.ylabel('Battito Cardiaco')
plt.show()

#data.columns
```
